```r
plot(df$age,df$charges,
    pch = 16,
    cex = 1.3,
    col = 'grey',
    xlab = 'AGE',
    ylab = 'CHARGE')
```

```r
lm(df$charges~df$age)
```

```r
plot(df$age,df$charges,
    pch = 16,
    cex = 1.3,
    abline(3165.9  , 257.7),
    col = 'grey',
    xlab = 'AGE',
    ylab = 'CHARGE')
```

```r
plot(df$bmi,df$charges,
    pch = 16,
    cex = 1.3,
    col = 'grey',
    xlab = 'BMI',
    ylab = 'CHARGE')
```

```r
lm(df$charges~df$bmi)
```

```r
plot(df$bmi,df$charges,
    pch = 16,
    cex = 1.3,
    col = 'grey',
    abline(1192.9 , 393.9),
    xlab = 'BMI',
    ylab = 'CHARGE')
```
