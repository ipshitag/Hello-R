# Regression

Loading Data and looking at its contents

````r
head(mtcars)
````

Correlation of different variables with respect to ```mpg```

````r
cor(mtcars,mtcars['mpg'])
````

We can see that, ```cyl```,```disp```and```wt``` are highly correlated with ```mpeg```

````r
# turning off warnings
options(warn=-1)
````

Looking at regression plots of the different variables with respect to ```mpeg```

````r
require(ggplot2)
require(reshape2)
mtcars2 = melt(mtcars, id.vars='mpg')
ggplot(mtcars2) +
  geom_jitter(aes(value,mpg, colour=variable),) + geom_smooth(aes(value,mpg, colour=variable), method=lm, se=FALSE) +
  facet_wrap(~variable, scales="free_x") +
  labs(x = "Percentage cover (%)", y = "Number of individuals (N)")
````

## Model 1

### With ```disp```,```cyl``` and ```wt```

````r
model1 <- lm(mtcars$mpg~mtcars$disp+mtcars$cyl+mtcars$wt)
````

````r
summary(model1)
````

### Results:

We can see, that our model gave us $R^2$ of $83.26%$

<hr>

## Model 2

### With ```disp```,```hp```,```drat``` and ```wt```

````r
model2 <- lm(mtcars$mpg~mtcars$disp+mtcars$hp+mtcars$drat+mtcars$wt)
````

````r
summary(model2)
````

### Results:

We can see, that our model gave us $R^2$ of $81.36%$

<hr>

## Model 3

### With ```hp```,```wt``` and ```cyl```

````r
model3 <- lm(mtcars$mpg~mtcars$hp+mtcars$wt+mtcars$cyl)
````

````r
summary(model3)
````

### Results:

We can see, that our model gave us $R^2$ of $82.63%$
